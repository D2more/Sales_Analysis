```python
#Cleaning of the data
#From the operations that was performed, some errors were discovred and we start by dropping all the NAN rows
#Drop Unamed column

df.drop(['Unnamed: 0'], axis = 1,inplace = True)
```

```python
df.head()
```

```python

#Find the NAN

nan_df = df.isna().any(axis=1)
display(nan_df.head())

df.dropna(how='all')
df.head()
```

```python
#Get the unique products
df['Product'].unique()
```

```python
df = df[df['Order Date'].str[0:2]!='Or']
```

```python
#Data Type Conversion for Numeric Columns:

df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df.head()
```

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int32')
df.head()
```

```python
#Add city column

def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
df.head()
```

```python
#Data Exploration!


#Question 1: What was the best month for sales? How much was earned that month?

df['Sales'] = df['Quantity Ordered'].astype('int') * df['Price Each'].astype('float')
df.groupby(['Month']).sum()
```

```python
#Question 2: What city sold the most product?
df.groupby(['City']).sum()
```

```python
#Product categories analysis

#We start by finding the unique(distinct) products from the dataset
unique_products = df["Product"].unique()
unique_products = pd.Series(df["Product"].unique())
unique_products.head()
```

```python
#Now we need to figure out the top Products by Quantity Ordered:

product_quantities = df[["Product", "Quantity Ordered"]].groupby("Product").sum()
product_quantities.head()
```

```python
#Quantity Ordered per Product - Bar Chart

df.groupby('Product')['Quantity Ordered'].sum().plot.bar(title = 'Quantity Ordered per Product',figsize =(4,5))
```

```python
#We can see 5 most sold products by quantity orders. We will divide a number of sold products on mean price.
#Product Performance Metric Calculation

products = df["Product"].unique()
df = {product: product_quantities.loc[product, "Quantity Ordered"] / df[df["Product"] == product]["Price Each"].mean() 
      for product in products if product in product_quantities.index}
df = pd.Series(df).sort_values(ascending=False)
df.head()
```

```python
#Monthly Sales

df.groupby('Month')['Sales'].sum().plot.bar(title = 'Monthly Sales',figsize = (4,5))
plt.savefig('/Users/YourUsername/Desktop/your_chart.png')
```

```python
#Top 6 Cities with the largest sales

city_sales = df.groupby('City')['Sales'].sum().nlargest(6)
city_sales
```

```python
#Top 6 cities with most sales

plt.figure(figsize = (8,8))
plt.title('Top 6 Cities with Largest Sales')
plt.pie(city_sales,autopct = '%.2f',labels = city_sales.index)
plt.show()
```

```python
df.groupby('Hour')['Sales'].sum().plot.line(marker='o', title='Sales per Hour', figsize=(12, 5))

plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.xticks(range(1, 25))  # Explicitly set x-axis ticks for each hour
#plt.grid(True)
plt.show()
```
